```python
# Defining a function to plot the ROC curve
def auc_roc_plot(y_test, y_pred_proba):
    
    '''
    Parameters: 
    - y_test: Series type object. Actual target values in the test data.

    - y_pred_proba: Series type object. Predicted probabilities of the target values.
    
    Returns: 
    Creates a plot of the receiver operating characteristic (ROC) curve
    '''
    
    fprs, tprs, thresholds = roc_curve(y_test, y_pred_proba)
    roc_auc = roc_auc_score(y_test, y_pred_proba)

    plt.figure()
    plt.plot(fprs, tprs, lw=2, color='navy', label='AUC = %0.2f' % roc_auc) #
    plt.plot([0, 1], [0, 1], color='red', lw=2, linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve')
    plt.legend()
    plt.show()

    print(f'Area under curve (AUC):{roc_auc}')
    print()
```

```python
# Defining a function to create a performance report for each models
def performance_report(y_test, y_pred, y_pred_proba):
    
    '''
    Parameters: 
    - y_test: Series type object. Actual target values in the test data.

    - y_pred: Series type object. Predictions of the target values.
    
    - y_pred_proba: Series type object. Predicted probabilities of the target values.

    Returns: 
    Does not return any results instead it displays the confusion matrix of y_test vs y_pred.
    It also calls the auc_roc_plot function to plot the receiver operating characteristic curve
    '''
    
    from sklearn.metrics import confusion_matrix
    
    cnf_matrix = confusion_matrix(y_test, y_pred)

    cf_df = pd.DataFrame(
        cnf_matrix, 
        columns=["Predicted Class 0", "Predicted Class 1"],
        index=["True Class 0", "True Class 1"]
    )
    display(cf_df)
    
    print(classification_report(y_test, y_pred))

    auc_roc_plot(y_test, y_pred_proba)
```

```python
# Defining a function to remove outliers based on IQR
def remove_outliers(df, columns):
    
    '''
    Parameters: 
    - df: DataFrame type object.

    - columns: List type object. Column names to remove outliers for.
    
    Returns: 
    Cleaned dataframe without outliers in the columns specified. Outliers are defined here as any values greater
    than Q3 plus 1.5*IQR or smaller than than Q1 less 1.5*IQR.
    '''
    
    df_cleaned = df.copy()
    
    for column in columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        df_cleaned = df_cleaned[~((df_cleaned[column] < (Q1 - 1.5 * IQR)) |(df_cleaned[column] > (Q3 + 1.5 * IQR)))]
    return df_cleaned
```

```python
def get_metrics(y_test, y_pred, y_pred_proba):
    
    '''
    Parameters: 
    - y_test: Series type object. Actual target values in the test data.

    - y_pred: Series type object. Predictions of the target values.
    
    - y_pred_proba: Series type object. Predicted probabilities of the target values.

    Returns: 
    A tuple containing the macro averaged Precision, Recall, F1 and AUC scores
    '''
    
    precision = precision_score(y_test, y_pred, average = 'macro')
    recall = recall_score(y_test, y_pred, average = 'macro')
    f1 = f1_score(y_test, y_pred, average = 'macro')
    roc_auc = roc_auc_score(y_test, y_pred_proba)
    scores = (precision, recall, f1, roc_auc)
    
    return scores
```

## Feature Selection

To begin feature selection, we can start with using hypothesis tests to confirm any statistical difference between applicants who defaulted vs those who repaid their loans:

### Statistics & Hypothesis testing

Using the cleaned data above, we can perform some hypothesis tests. Starting with the categorical data, we can implement the chi-squared test of independence assuming the null & alternative conditions below:

$H_0$: There is no relationship between loan status and the categorical variables. i.e both are independent.

$H_1$: There is some relationship between loan status and the categorical variables i.e both are not independent.

```python
# Get p_value from the chi2 test

cat_columns = df.select_dtypes('object').columns
for column in cat_columns:
    chi2_results = stats.chi2_contingency(pd.crosstab(df['loan_status'],df[column]))
    p_value = chi2_results[1]
    
    print(f"p-value for a chi2 test between loan status and {column} is {round(p_value,4)}\n")
```

Based on the p_values obtained above, all categorical features are important in predicting loan default except from:
* emp_title, full address and state of residence

The p_value obtained for those columns are much higher than 0.05, hence we fail to reject the null hypothesis. Instead, we can conclude that employee title, full address and state of the applicant are independent of loan_status.

Next, we can use a t-test to compare if the numeric columns are statistically significantly different across the defaulted and repaid classes:

Hypothesis Test:

<center> $H_0$: $\mu_{Default} = \mu_{Repaid} $ $\;\; vs \;\;$  $H_1$: $\mu_{Default} \neq \mu_{Repaid} $ </center>

```python
# T-test on the numeric data
num_columns = df.select_dtypes("number").columns
for column in num_columns[:]:
    
    default_df = df.loc[df['loan_status'] == 1][column]
    repaid_df = df.loc[df['loan_status'] == 0][column]
    
    p_value = stats.ttest_ind(default_df, repaid_df)[1]
    print(f"{column} by loan_status\nt-test p-value:{p_value}\n")
    plt.show()
```

The p-values are all low for each of the numeric columns as shown above. Since the p-values are < 0.05, we can safely reject the null hypothesis and conclude that the means of each column are statistically different across the two classes(default vs repaid). In other words, loan_status is dependent of each of the numeric columns.

### Variance Inflation Factor

To avoid creating an unstable and unreliable model, we need to assess multicollinearity among the predictor variables. By looking at the features with high variance inflation factor, we can exclude unnecessary features in our dataset that have strongly correlations with other predictor variables. By general convention, any VIF value higher than 5 indicates high collinearity. We will use 5 as the threshold moving forward:

## Model Training and Evaluation

Given that this is an imbalanced dataset, accuracy score would not be a good metric to use for benchmarking. That is because it can give a false sense of model success by simply performing well on the majority class although it performs poorly on the minority class. Instead, I will use AUC to compare models as it provides an aggregate measure of the each model's ability to properly seperate  between the positive and negative classes across varying probability thresholds. It considers both false positive rate and true positive rate:

$$TPR = \frac{TP}{TP+FN}$$

$$FPR = \frac{FP}{FP+TN}$$

The True Positive Rate (TPR) measures what proportion of clients our model classifies as defaulters are actually members of the defaulters class. It tells us how well our model is at classifying the true loan defaulters. On the other hand, the False Positive Rate (FPR) will be the proportion of all credit-worthy clients that we incorrectly identified defaulters. In other words, that is the proportion of clients we missed out on serving.

To strike the ideal balance between mitigating our loss and maximizing our profits, we need to identify the highest amount of true loan defaulters while providing loans to the maximum amount of deserving clients. I will optimize for AUC and use that as the principal metric for model comparison.

#### Optimizing C Using 3-Fold Cross Validation

Next we can try to optimize the Logistic Regression C parameter by evaluating the model's performance for varying values between .0001 to 1000. We will vary the scoring metric between F1, recall, precision and AUC in order to observe how the performance varies with C:

```python
import warnings
warnings.filterwarnings('always')
```

```python
cv_f1_scores = []
cv_recall_scores = []
cv_precision_scores = []
cv_auc_scores = []

C_range = np.array([.0001,.001,.01,0.1,\
                1,10,100,1000,1000])

for c in C_range:
    
    logreg_model = LogisticRegression(random_state=123, C = c, n_jobs=-1, max_iter=2000)
    
    cv_f1_score = np.mean(cross_val_score(logreg_model, X_train_scaled, y_train, cv = 3, scoring = 'f1_macro'))
    cv_recall_score = np.mean(cross_val_score(logreg_model, X_train_scaled, y_train, cv = 3, scoring = 'recall_macro'))
    cv_precision_score = np.mean(cross_val_score(logreg_model, X_train_scaled, y_train, cv = 3, scoring = 'precision_macro'))
    cv_auc_score = np.mean(cross_val_score(logreg_model, X_train_scaled, y_train, cv = 3, scoring = 'roc_auc'))


    cv_f1_scores.append(cv_f1_score)
    cv_recall_scores.append(cv_recall_score)
    cv_precision_scores.append(cv_precision_score)
    cv_auc_scores.append(cv_auc_score)
```

```python
plt.figure(figsize = (12,5))
plt.plot(C_range, cv_f1_scores, label="Cross Validation - F1",marker='.', color = "red")
plt.plot(C_range, cv_recall_scores, label="Cross Validation - Recall",marker='.', color = "blue")
plt.plot(C_range, cv_precision_scores, label="Cross Validation -Precision",marker='.', color = "yellow")
plt.plot(C_range, cv_auc_scores, label="Cross Validation Score - AUC",marker='.', color = "black")
plt.legend()
plt.xlabel('C Parameter')
plt.ylabel('Cross Validation Average Scores')
plt.xscale('log')
plt.xticks(np.array([.0001,.001,.01,0.1,1,10,100,1000,1000]))
plt.show();
```

```python
C_range[np.argmax(cv_auc_scores)]
```

From the plot above, Logistic Regression model with C = 10 maximizes the AUC, macro averaged Recall and F1 score. Although there is a slight trade off with Precision score. Using 10 as the optimal C value, we can see how the chosen model performs on the test data:

#### Hyperparameter Tuning Logistic Regression using GridSearchCV

Using GridSearch, we can also optimize multiple parameters in the Logistic Regression model at once. I will vary:
* Penalty Used: L1 or L2
* Dimensionality reduction: With & without PCA. When dimensions are reduced, vary amount of variance retained between 80% to 95%
* Vary C parameter between .0001 and 1000

```python
cachedir = mkdtemp()

# Set up a pipeline
log_pipeline = Pipeline([('scaler', StandardScaler()), ('dim_reducer', PCA()), ('model', LogisticRegression())], memory = cachedir)

C = [.0001,.001,.01,0.1,1,10,100,1000,1000]

log_param_grid = [
    
    # L1 penalty without PCA
    {'scaler': [StandardScaler(), MinMaxScaler()],
     'dim_reducer': [None],
     'model': [LogisticRegression(penalty='l1', random_state=123, max_iter=2000, n_jobs=-1)],
     'model__C': C},
    
     # L2 without PCA
    {'scaler': [StandardScaler(), MinMaxScaler()],
     'dim_reducer': [None],
     'model': [LogisticRegression(penalty='l2', solver='lbfgs', random_state=123, max_iter=2000, n_jobs=-1)],
     'model__C': C},
    
    # L1 penalty with PCA
    {'scaler': [StandardScaler(), MinMaxScaler()],
     'dim_reducer': [PCA()],
     'dim_reducer__n_components': [0.95, 0.85],
     'model': [LogisticRegression(penalty='l1', random_state=123, max_iter=2000, n_jobs=-1)],
     'model__C': C},
    
    # L2 with PCA
    {'scaler': [StandardScaler(), MinMaxScaler()],
     'dim_reducer': [PCA()],
     'dim_reducer__n_components': [0.95, 0.85],
     'model': [LogisticRegression(penalty='l2', solver='lbfgs', random_state=123, max_iter=2000, n_jobs=-1)],
     'model__C': C}
]
```

```python
# Instantiating & fitting the grid search
gridsearch_logreg = GridSearchCV(log_pipeline, param_grid = log_param_grid, cv=3, n_jobs=-1, verbose=1)

fitted_log_gs = gridsearch_logreg.fit(X_train, y_train)
fitted_log_gs.best_estimator_
```

The best logistic regression model to use based on the grid search optimization above is LogisticRegression(C=0.01, max_iter=2000, n_jobs=-1, random_state=123) preprocessed with StandardScaler and no PCA dimensionsion reduction applied. Let's see how this model performs on the test data below:

#### Optimizing max_depth Using 3-Fold Cross Validation

```python
cv_f1_scores = []
cv_recall_scores = []
cv_precision_scores = []
cv_auc_scores = []

max_depth = np.arange(1,22,2)

for depth in max_depth:
    
    tree_model = DecisionTreeClassifier(random_state = 123, splitter = 'best', max_depth = depth)
    tree_model.fit(X_train_scaled, y_train)
    
    cv_f1_score = np.mean(cross_val_score(tree_model, X_train_scaled, y_train, cv = 3, scoring = 'f1_macro'))
    cv_recall_score = np.mean(cross_val_score(tree_model, X_train_scaled, y_train, cv = 3, scoring = 'recall_macro'))
    cv_precision_score = np.mean(cross_val_score(tree_model, X_train_scaled, y_train, cv = 3, scoring = 'precision_macro'))
    cv_auc_score = np.mean(cross_val_score(tree_model, X_train_scaled, y_train, cv = 3, scoring = 'roc_auc'))


    cv_f1_scores.append(cv_f1_score)
    cv_recall_scores.append(cv_recall_score)
    cv_precision_scores.append(cv_precision_score)
    cv_auc_scores.append(cv_auc_score)
```

####  Optimizing n_estimators using 3-Fold Cross Validation

Still assuming an ideal max_depth of 15 for our decision trees, we can find the ideal number of estimators/trees to use in our random forest model as shown below:

```python
cv_f1_scores = []
cv_recall_scores = []
cv_precision_scores = []
cv_auc_scores = []

n_range = range(5, 100, 10)

for n in n_range:
    RForest_model = RandomForestClassifier(n_estimators = n, max_depth = 15)
    RForest_model.fit(X_train_scaled, y_train)
    
    cv_f1_score = np.mean(cross_val_score(RForest_model, X_train_scaled, y_train, cv = 3, scoring = 'f1_macro'))
    cv_recall_score = np.mean(cross_val_score(RForest_model, X_train_scaled, y_train, cv = 3, scoring = 'recall_macro'))
    cv_precision_score = np.mean(cross_val_score(RForest_model, X_train_scaled, y_train, cv = 3, scoring = 'precision_macro'))
    cv_auc_score = np.mean(cross_val_score(RForest_model, X_train_scaled, y_train, cv = 3, scoring = 'roc_auc'))


    cv_f1_scores.append(cv_f1_score)
    cv_recall_scores.append(cv_recall_score)
    cv_precision_scores.append(cv_precision_score)
    cv_auc_scores.append(cv_auc_score)

plt.figure(figsize = (12,5))
plt.plot(n_range, cv_f1_scores, label="Cross Validation - F1",marker='.', color='green')
plt.plot(n_range, cv_recall_scores, label="Cross Validation - Recall",marker='.', color='black')
plt.plot(n_range, cv_precision_scores, label="Cross Validation -Precision",marker='.', color='blue')
plt.plot(n_range, cv_auc_scores, label="Cross Validation Score - AUC",marker='.', color='red')
plt.legend()
plt.xlabel('Number of estimators')
plt.ylabel('Cross Validation Average Scores')
plt.xticks(n_range)
plt.grid()
plt.show()
```

Looking at how the average cross validation scores changed with increasing number of estimators shown above, we can see that the AUC increases sightly between n_estimators of 5 and 25. After this, we do not gain any better classification of clients as the AUC seems to plateau after n_estimators of 25. Next, let us see how the model performs using 25 trees:

Logistic Regression model provided the best performance with ROC_AUC score of 0.9043. It identifies the true loan defaulters with 96% precision while having 88% precision on the loan repaid class. Compared to the baseline accuracy of 80% if the model predicted clients would pay 100% of the time, we increased prediction accuracy by 9% using this model. I believe this model can be very useful for a wide range of institutions including banks and financial institutions, peer-to-peer lenders, credit rating agencies, insurance companies, etc. These companies can improve their credit risk management practices while maximizing their profits.
