#### Correlations

Converting the target column to a binary variable where charged off loans = class 1 and paid off = 0:

```python
df["loan_status"] = np.where(df["loan_status"] == 'Charged Off', 1, 0)
```

```python
numeric_df = df[list(df.select_dtypes("number").columns)]
mask = np.triu(numeric_df.corr(), 1)
plt.figure(figsize=(15,7))
sns.heatmap(numeric_df.corr(), vmax=1, vmin=-1, annot=True, mask=mask, cmap=sns.diverging_palette(220, 20, as_cmap=True))
```

The correlation between the rest of the variables:

- 'installment' and 'loan amount' have a strong positive correlation with a coefficient of 0.95. It can be expected that the higher the loan amount, the higher the number of installments needed to repay the loan. To avoid multicolinearity while running the Logistic Regression model, I will drop the 'installment' column from the training dataset as the 'loan_amount' already provides insight into the loan/repayment amount.

- 'open_acc' and 'total_acc' are positively correlated with a coefficient of 0.68. Both columns provide insight into how many accounts an individual has. It may not be essential to keep both columns during the modelling phase but that can be tested later on.

- 'pub_rec' and 'pub_rec'_bankruptcies also have a strong positive correlation with a coefficient of 0.70. Both columns provide insight into how many derogatory events/delinquencies an applicant has has in the past. Similarly, one of these columns could be dropped during the modelling phase.

- the higher the number of total accounts an individual has, the higher their number of mortgage accounts is

- 'annual_inc' has a weak positive correlation with 'loan_amnt' with a coefficient of 0.34. From this dataset, we can see that the higher the customer's annual income, the higher their loan amount was.

Relative to the target variable:

- interest rate has the highest positive correlation with the loan_status. The higher the interest rate on the loan, the more likely the applicant will default. Although it is a weak positive correlation with a coefficient of 0.25.
